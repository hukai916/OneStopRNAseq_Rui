```python
with open("clades_extended.csv", "w") as oup:
    def recursiveChildren(x, stk):
        if "childGenerator" in dir(x):
            ncnt = 0
            for child in x.childGenerator():
                name = getattr(child, "name", None)
                txt = ''
                if name is not None:
                    txt = child.text.split('\n')[0]
                    ncnt += 1
                else:
                    continue
                recursiveChildren(child, stk + [txt]) if len(txt)>0 else recursiveChildren(child, stk)
        if ncnt == 0:
            tmp = (stk + [stk[-1]] * (4-len(stk)))
            tmp = tmp[::-1]
            print(",".join(tmp), file=oup)
    recursiveChildren(soup.ul.small, [])
```
